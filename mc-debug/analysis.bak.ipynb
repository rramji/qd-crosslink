```python
x = lammps_mol[4]['Step']
fig, axs = plt.subplots(nrows=4, ncols=3,figsize=(20,20))
plt.subplots_adjust(hspace=0.5)
fig.suptitle("Comparison of LAMMPS thermo data", fontsize=18, y=0.95)
dat = lammps_mol[4].columns[1:]
for col, ax in zip(dat, axs.ravel()):
    
    y1 = lammps_mol[4][col]
    y2 = data_parm[4][col]
    y3 = input_parm[4][col]
    
    ax.set_title(col)
    ax.plot(x,y1, color='blue')
    ax.plot(x,y2, color='black')
    ax.plot(x,y3, color='red')
```
